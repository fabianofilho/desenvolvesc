```python
import pandas as pd
```

```python
df = pd.read_csv('marcacoes_2018-v1.tab.gz', compression='gzip', header=0, sep='\t',)
```

```python
df.info()
```

```python
df.head()
```

```python
df = df[df['_DATE_DIFF_data_exec_x_data_solic'] > 0] 
df = df[df['_DATE_DIFF_data_exec_x_data_autoriz'] > 0] 

# Separar o X com caracteristicas influenciadoras do Y
X = df[['ibge_munic_resid',
        'classif_risco',
        'sexo',
        '_DATE_DIFF_now_x_data_nasc',
        'desc_proced'        
       ]]
y = df.confirmacao

# Classificar colunas em 0 e 1
X = pd.get_dummies(X)

# Evitar dummie trap no sexo
X = X.drop('sexo_MASCULINO', 1)

# Transformar codigo para string
X['ibge_munic_resid'] = X['ibge_munic_resid'].astype(str)

# Reclassificar colunas em 0 e 1
X = pd.get_dummies(X)

X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
```

```python
# TREINANDO O MODELO
dt.fit(X_train, y_train)

# FAZENDO PREDIÇÕES DO MODELO
y_pred = dt.predict(X_test)
```

```python
# evaluate predictions
from sklearn.metrics import classification_report, accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
print(classification_report(y_test, y_pred))
```
