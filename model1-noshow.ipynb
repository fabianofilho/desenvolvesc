```python
import pandas as pd
```

```python
df = pd.read_csv('marcacoes_2018-v1.tab.gz', compression='gzip', header=0, sep='\t',)
```

```python
df.info()
```

```python
df.head()
```

```python
df = df[df['_DATE_DIFF_data_exec_x_data_solic'] > 0] 
df = df[df['_DATE_DIFF_data_exec_x_data_autoriz'] > 0] 
```

```python
df.columns.tolist()
```

```python
df.isnull().sum()
```

```python
X = df[['ibge_munic_resid',
        'classif_risco',
        'sexo',
        '_DATE_DIFF_now_x_data_nasc',
        'desc_proced'        
       ]]
y = df.confirmacao
```

```python
X = pd.get_dummies(X)
```

```python
# Evitar dummie trap no sexo
X = X.drop('sexo_MASCULINO', 1)

# Transformar codigo para string
X['ibge_munic_resid'] = X['ibge_munic_resid'].astype(str)
X.head()
```

```python
X = pd.get_dummies(X)
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
rf.fit(X_train, y_train)
```

```python
p = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, p))
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
# Create and fit selector
selector = SelectKBest(f_classif, k=5)
selector.fit(X, y)
```

```python
# Get columns to keep
cols = selector.get_support(indices=True)
```
